```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Example DataFrame
data = {'SquareFootage': [800, 1000, 1200, 1500, 1800, 2000, 2300],
        'Price': [150000, 200000, 250000, 270000, 300000, 330000, 350000]}
df = pd.DataFrame(data)

# Scatter plot with regression line
sns.regplot(x='SquareFootage', y='Price', data=df, ci=None)
plt.xlabel('Square Footage')
plt.ylabel('Price')
plt.title('Linear Regression of Housing Prices')
plt.show()

import numpy as np

# Calculating predicted prices and residuals
df['PredictedPrice'] = df['SquareFootage'] * 150 + 100000  # Simplified linear model
df['Residuals'] = df['Price'] - df['PredictedPrice']

# Residual plot
plt.scatter(df['PredictedPrice'], df['Residuals'])
plt.axhline(y=0, color='r', linestyle='--')
plt.xlabel('Predicted Price')
plt.ylabel('Residuals')
plt.title('Residual Plot')
plt.show()


```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Step 1: Create a DataFrame with initial housing data
data = {
    'SquareFootage': [800, 1000, 1200, 1500, 1800, 2000, 2300],
    'Price': [150000, 200000, 250000, 270000, 300000, 330000, 350000]
}
df = pd.DataFrame(data)

# Step 2: Introduce an outlier (corrected)
df.loc[len(df)] = [2500, 600000]  # Adding an outlier point with correct format

# Step 3: Scatter plot with regression line including the outlier
plt.figure(figsize=(8, 6))
sns.regplot(x='SquareFootage', y='Price', data=df, ci=None)
plt.xlabel('Square Footage')
plt.ylabel('Price')
plt.title('Linear Regression with Outliers')
plt.show()
```
